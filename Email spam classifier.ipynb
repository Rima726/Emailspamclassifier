```python
import os
from collections import Counter
```

```python
files=os.listdir('email')
```

```python
for i in files:
    print(i)
```

```python
emails=[]
for i in files:
    emails.append('email/'+i)
```

```python
emails
```

```python
words=[]
for i in emails:
    f=open(i,encoding='latin-1')
    words.extend(f.read().split(" "))
```

```python
len(words)
```

```python
for i in range(len(words)):
    if not words[i].isalpha():
        words[i]=""
```

```python
word_dict=Counter(words)
```

```python
word_dict.most_common(1)
```

```python
del word_dict['']
```

```python
len(word_dict)
```

```python
new_word_dict=word_dict.most_common(3000)
```

```python
new_word_dict
```

```python
features=[]
labels=[]
for i in emails:
    if 'spam'in i:
        labels.append(1)
    if 'ham'in i: 
        labels.append(0)
        
    f=open(i,encoding='latin_1')
    text=f.read()
    data=[]
    for j in new_word_dict:
        data.append(text.count(j[0]))
    features.append(data)   
        
```

```python
import numpy as np
```

```python
X=np.array(features)
y=np.array(labels)
```

```python
print(X.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.naive_bayes import MultinomialNB
clf=MultinomialNB()
```

```python
mail="""
Hi you have won an Apple Iphone 13
Please share your phone number and bank account details
so that we can contact you
"""
```

```python
clf.fit(X_train,y_train)
```

```python
input_mail=[]
for i in new_word_dict:
    input_mail.append(mail.count(i[0]))
```

```python
clf.predict(np.array(input_mail).reshape(1,3000))
```

```python
y_pred=clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
